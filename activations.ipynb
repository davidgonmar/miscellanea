```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

def leaky_relu(x, alpha=0.3):
    return np.where(x > 0, x, alpha * x)

def leaky_relu_derivative(x, alpha=0.3):
    return np.where(x > 0, 1, alpha)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1, elu(x, alpha) + alpha)

def selu(x, lambda_=1.0507, alpha=1.67326):
    return lambda_ * np.where(x > 0, x, alpha * (np.exp(x) - 1))

def selu_derivative(x, lambda_=1.0507, alpha=1.67326):
    return lambda_ * np.where(x > 0, 1, alpha * np.exp(x))

def swish(x):
    return x * sigmoid(x)

def swish_derivative(x):
    s = sigmoid(x)
    return s + x * s * (1 - s)

def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

def gelu_derivative(x):
    tanh_term = np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3)))
    return 0.5 * (1 + tanh_term) + (0.5 * x * (1 - tanh_term**2) * (np.sqrt(2 / np.pi) * (1 + 3 * 0.044715 * x**2)))

def softplus(x):
    return np.log1p(np.exp(x))

def softplus_derivative(x):
    return sigmoid(x)

x = np.linspace(-10, 10, 400)

functions = [
    (sigmoid, sigmoid_derivative, 'Sigmoid'),
    (tanh, tanh_derivative, 'Tanh'),
    (relu, relu_derivative, 'ReLU'),
    (leaky_relu, leaky_relu_derivative, 'Leaky ReLU'),
    (elu, elu_derivative, 'ELU'),
    (selu, selu_derivative, 'SELU'),
    (swish, swish_derivative, 'Swish'),
    (gelu, gelu_derivative, 'GELU'),
    (softplus, softplus_derivative, 'Softplus')
]

fig, axes = plt.subplots(len(functions), 2, figsize=(14, 36))

for i, (func, derivative, name) in enumerate(functions):
    axes[i, 0].plot(x, func(x), label=f'{name}', color='blue')
    axes[i, 0].set_title(f'{name} Function')
    axes[i, 0].legend()
    axes[i, 0].grid(True)

    axes[i, 1].plot(x, derivative(x), label=f'{name} Derivative', color='red')
    axes[i, 1].set_title(f'{name} Derivative')
    axes[i, 1].legend()
    axes[i, 1].grid(True)

plt.tight_layout()
plt.show()
```
